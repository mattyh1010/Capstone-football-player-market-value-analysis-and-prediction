# Data Cleaning

Due to prior expertise of data, was able to condence variables down by grouping similar columns together

```python
df["goalkeeping_ability"] = (df["Thr"] + df["TRO"] + df["Ref"] + df["Pun"] + df["1v1"] + df["Kic"] + df["Han"] + df["Ecc"] + df["Com"] + df["Cmd"])/10
```

```python
df = df.drop(['Thr', 'TRO', 'Ref', 'Pun', '1v1', 'Kic', 'Han', 'Ecc', 'Com', 'Cmd'], axis=1)
```

```python
df["speed"] = (df["Pac"] + df["Acc"] + df["Agi"])/3
df = df.drop(['Pac', 'Acc', 'Agi'], axis=1)
```

```python
df["aerial"] = (df["Aer"] + df["Hea"] + df["Jum"])/3
df = df.drop(['Aer', 'Hea', 'Jum'], axis=1)
```

```python
df["speed"] = df["speed"].round(1)
```

```python
df["aerial"] = df["aerial"].round(1)
```

```python
df["set_pieces"] = (df["Fre"] + df["Cor"] + df["Pen"])/3
df = df.drop(['Fre', 'Cor', 'Pen'], axis=1)
df["set_pieces"] = df["set_pieces"].round(1)
```

```python
df["shooting"] = (df["Fin"] + df["Lon"])/2
df = df.drop(['Fin', 'Lon'], axis=1)
df["shooting"] = df["shooting"].round(1)
```

```python
df["stamina"] = (df["Sta"] + df["Nat"] + df["Wor"])/3
df = df.drop(['Sta', 'Nat', 'Wor'], axis=1)
df["stamina"] = df["stamina"].round(1)
```

```python
df["strength"] = (df["Agg"] + df["Str"])/2
df = df.drop(['Agg', 'Str'], axis=1)
df["strength"] = df["strength"].round(1)
```

```python
df = df.rename(columns={"Ant" : "anticipation"})
```

```python
df = df.rename(columns={"Bal" : "balance", "Bra" : "bravery", "Cnt" : "concentration", "Cro" : "crossing", "Det" : "determination", "Ldr" : "leadership", "Tea" : "teamwork"})
```

```python
df["decision_making"] = (df["Dec"] + df["Cmp"])/2
df = df.drop(['Dec', 'Cmp'], axis=1)
df["decision_making"] = df["decision_making"].round(1)
```

```python
df["creativity"] = (df["Dri"] + df["Fla"] + df["Vis"])/3
df = df.drop(['Dri', 'Fla', 'Vis'], axis=1)
df["creativity"] = df["creativity"].round(1)
```

```python
df["technical"] = (df["Fir"] + df["Tec"] + df["Pas"] + df["OtB"])/4
df = df.drop(['Fir', 'Tec', 'Pas', "OtB"], axis=1)
df["technical"] = df["technical"].round(1)
```

```python
df["defending"] = (df["Mar"] + df["Tck"] + df["Pos"])/3
df = df.drop(['Mar', 'Tck', 'Pos'], axis=1)
df["defending"] = df["defending"].round(1)
```

# Modelling

```python
from sklearn.model_selection import train_test_split

import statsmodels.api as sm
import statsmodels.tools
```

Adding constant for Linear Regression

```python
df = sm.add_constant(df)
```

Creating feature columns

```python
feature_cols = list(df.columns)
feature_cols.remove("value")
feature_cols.remove("sqrt_value")
```

Creating X and y

```python
X = df[feature_cols]
y = df["sqrt_value"]
```

```python
X.head()
```

Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 20)
```

```python
feature_cols = ['const', 'age', 'preferred_foot',
                'wage', 'ca', 'teamwork', 'leadership', 'concentration',
                'goalkeeping_ability', 'aerial', 'set_pieces',
                'stamina', 'defending']
```

Creating Linear Regression Model

```python
lin_reg = sm.OLS(y_train, X_train[feature_cols])
results = lin_reg.fit()
results.summary()
```

Finding RMSE of train

```python
y_pred = results.predict(X_train[feature_cols])

rmse_train = statsmodels.tools.eval_measures.rmse(y_train, y_pred)

print(rmse_train)
```

Finding RMSE of test

```python
y_pred = results.predict(X_test[feature_cols])

rmse_test = statsmodels.tools.eval_measures.rmse(y_test, y_pred)

print(rmse_test)
```

Using delta method to back-transform RMSE

```python
approx_train_rmse = (y_train.mean()*2) * (rmse_train)
print(approx_train_rmse)
```

```python
approx_test_rmse = (y_test.mean()*2) * (rmse_test)
print(approx_test_rmse)
```

#Ridge Regression Model

```python
from sklearn.preprocessing import scale
from sklearn.linear_model import Ridge, RidgeCV
from sklearn.metrics import mean_squared_error
```

```python
feature_cols_ridge = ['const', 'age', 'preferred_foot',
                'wage', 'ca', 'pa', 'teamwork', 'leadership', 'determination', 'crossing', 'concentration',
                'bravery', 'balance', 'anticipation', 'goalkeeping_ability', 'speed', 'aerial', 'set_pieces',
                'shooting', 'stamina', 'strength', 'decision_making', 'creativity', 'technical', 'defending']
```

Ridge regression RMSE on train

```python
ridge2 = Ridge(alpha = 4)
ridge2.fit(X_train[feature_cols_ridge], y_train)
pred2 = ridge2.predict(X_train[feature_cols_ridge])
print(mean_squared_error(y_train, pred2))
print(np.sqrt(mean_squared_error(y_train, pred2)))
```

Ridge regression RMSE on test

```python
pred3 = ridge2.predict(X_test[feature_cols_ridge])
print(mean_squared_error(y_test, pred3))
print(np.sqrt(mean_squared_error(y_test, pred3)))
```

Back-transformed RMSE for ridge regression

```python
approx_train_rmse_ridge = (y_train.mean()*2) * (np.sqrt(mean_squared_error(y_train, pred2)))
print(approx_train_rmse_ridge)
```

```python
approx_test_rmse_ridge = (y_test.mean()*2) * (np.sqrt(mean_squared_error(y_test, pred3)))
print(approx_test_rmse_ridge)
```

Cross validation on Ridge regression alpha

```python
alphas = 10**np.linspace(10,-2,100)*0.5
ridgecv = RidgeCV(alphas = alphas, scoring = 'neg_mean_squared_error')
ridgecv.fit(X_train[feature_cols_ridge], y_train)
ridgecv.alpha_
```

Testing CV results

```python
ridge3 = Ridge(alpha = ridgecv.alpha_)
ridge3.fit(X_train[feature_cols_ridge], y_train)
print(mean_squared_error(y_train, ridge3.predict(X_train[feature_cols_ridge])))
print(np.sqrt(mean_squared_error(y_train, ridge3.predict(X_train[feature_cols_ridge]))))
```

```python
print(mean_squared_error(y_test, ridge3.predict(X_test[feature_cols_ridge])))
print(np.sqrt(mean_squared_error(y_test, ridge3.predict(X_test[feature_cols_ridge]))))
```

Back-transformed RMSE for CV Ridge Regression

```python
approx_train_rmse_ridge_cv = (y_train.mean()*2) * (np.sqrt(mean_squared_error(y_train, ridge3.predict(X_train[feature_cols_ridge]))))
print(approx_train_rmse_ridge_cv)
```

```python
approx_test_rmse_ridge_cv = (y_test.mean()*2) * (np.sqrt(mean_squared_error(y_test, ridge3.predict(X_test[feature_cols_ridge]))))
print(approx_test_rmse_ridge_cv)
```
